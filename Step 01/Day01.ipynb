## 1. 인공지능이 데이터를 이해하는 방식
인공지능 모델 내부에서 모든 데이터(이미지, 텍스트, 음성 등)는 **숫자의 묶음**인 **텐서(Tensor)**로 변환됩니다.

### 텐서의 종류
* **스칼라(Scalar):** $0D$ 텐서 / 하나의 숫자 값
* **벡터(Vector):** $1D$ 텐서 / 숫자의 나열 (1차원 배열)
* **행렬(Matrix):** $2D$ 텐서 / 숫자의 표 (2차원 배열)
* **텐서(Tensor):** $3D$ 이상의 텐서 / 행렬을 쌓아 올린 구조

---

## 2. 핵심 이론: 행렬 곱셈 (Matrix Multiplication)
AI 모델의 학습과 추론은 수많은 행렬 곱셈의 연속입니다.

### 💡 계산 규칙 및 조건
1. **차원 일치:** 앞 행렬($A$)의 **열(Column)** 개수와 뒤 행렬($B$)의 **행(Row)** 개수가 반드시 같아야 합니다.
2. **결과 차원:** $(n \times m)$ 행렬과 $(m \times p)$ 행렬을 곱하면 결과는 $(n \times p)$가 됩니다.

---

## 3. 구현: 행렬 곱셈기 (Pure Python)
라이브러리를 사용하지 않고, 동작 원리 파악을 위해 직접 구현해 봅니다.

```python
def multiply_matrices(A, B):
    # 1. 행렬의 크기(Shape) 파악
    # len(A)는 행(세로)의 개수, len(A[0])은 열(가로)의 개수입니다.
    rows_A, cols_A = len(A), len(A[0])
    rows_B, cols_B = len(B), len(B[0])

    # 2. 곱셈 가능 여부 체크 (중요!)
    # 앞 행렬의 가로(cols_A)와 뒤 행렬의 세로(rows_B)가 같아야만 짝이 맞습니다.
    if cols_A != rows_B:
        raise ValueError("앞 행렬의 가로와 뒤 행렬의 세로 개수가 같아야 합니다.")

    # 3. 결과 행렬 틀 만들기
    # 결과 행렬은 (A의 세로 x B의 가로) 크기가 됩니다.
    # 일단 모든 칸을 0으로 채워둔 빈 표를 만든다고 생각하세요.
    result = [[0 for _ in range(cols_B)] for _ in range(rows_A)]

    # 4. 3중 for문: 행렬 곱셈의 핵심 (계산기 가동)
    
    # [첫 번째 for문] A행렬의 '몇 번째 줄(행)'을 계산할지 정합니다.
    for i in range(rows_A):
        
        # [두 번째 for문] B행렬의 '몇 번째 칸(열)'에 결과를 적을지 정합니다.
        for j in range(cols_B):
            
            # [세 번째 for문] 실제로 '가로 x 세로'를 해서 더하는 과정입니다.
            # k는 A의 가로로 이동하면서 동시에 B의 세로로 내려가는 '인덱스' 역할을 합니다.
            for k in range(cols_A):
                # A[i][k]: i번째 줄에서 옆으로(k) 이동하며 숫자 선택
                # B[k][j]: j번째 칸에서 아래로(k) 내려가며 숫자 선택
                # 이 둘을 곱해서 결과 행렬의 (i, j) 자리에 계속 누적해서 더합니다.
                result[i][j] += A[i][k] * B[k][j]
    
    return result

# 테스트 실행
mat_a = [[1, 2], [3, 4]]
mat_b = [[5, 6], [7, 8]]

try:
    res = multiply_matrices(mat_a, mat_b)
    print(f"결과 행렬: {res}")
except ValueError as e:
    print(e)
    
```

## 4. 원인 분석 및 재발 방지 (Troubleshooting)
* **에러:** `IndexError` 또는 `ValueError`
* **원인:** 모델 레이어 간의 데이터 차원(Shape)이 맞지 않을 때 발생.
* **해결 팁:** 연산 전후로 `shape`를 확인하는 습관을 들여야 합니다. 
* **최적화 방향:** 파이썬의 `for`문은 매우 느리므로, 실무에서는 C로 최적화된 **NumPy**를 사용하여 **벡터화 연산(Vectorization)**을 수행해야 합니다.
