# Chapter 06. TensorFlow.js 웹 서비스 배포

이제 AI 모델은 서버에만 머물지 않습니다. 사용자의 웹 브라우저에서 직접 모델을 돌려보며, GPU 가속을 활용하는 현대적인 AI 웹 서비스를 구축합니다.

### 1. 브라우저 환경의 모델 로딩 (Step 1)
- **목표**: TensorFlow.js 라이브러리를 이해하고 모델을 브라우저 메모리에 로드.
- **Deep Dive**: WebGL 가속을 통한 고속 연산 원리와 서버 추론 vs 클라이언트 추론 비교.
- **Engineering**: 모델 가중치 데이터의 비동기 로딩 최적화 및 메모리 관리.

### 2. 실시간 객체 탐지 및 UI/UX 구현 (Step 2)
- **목표**: 브라우저 화면에 예측 결과(Bounding Box 등)를 실시간으로 렌더링.
- **Deep Dive**: Web Worker를 이용한 메인 스레드 비차단 연산(UI 프리징 방지).
- **Engineering**: 성능 유지를 위한 프레임 드랍(Frame Drop) 방지 전략.

# Chapter 06-1. 웹에서의 AI 연산 (In-Browser Inference)

TensorFlow.js를 사용하면 Python 환경 설치 없이도 웹 브라우저만으로 딥러닝 모델을 실행할 수 있습니다. 

> **Key Point**: 브라우저의 그래픽 카드 자원(WebGL)을 사용하여, 서버 비용을 들이지 않고도 사용자 기기에서 즉시 AI 연산을 수행합니다.
